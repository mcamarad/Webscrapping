```python
#Importing packages
from selenium import webdriver
```

```python
webs = ["https://www.maths.ox.ac.uk/people/luis.alday", 
        "https://selbornechambers.co.uk/our-members/barristers/", 
        "https://www.crick.ac.uk/research/find-a-researcher/lucia-prieto-godino"]
```

```python
#This requires Google Chrome installed in the computer. Change the path to the chromedriver needed.
driver = webdriver.Chrome('chromedriver.exe')

profiles = []
```

```python
driver.get(webs[0])

ids = driver.find_elements_by_xpath("//*[contains(@id,'user-')]")[2]
x = ids.get_attribute("id")

name_element = driver.find_elements_by_xpath('/html/head/title')[0]
name = name_element.get_attribute("text").split("|")[0]
telephone_element = driver.find_elements_by_xpath('//*[@id="'+ x +'"]/div[1]/p')[0]
telephone = telephone_element.text
occupation_element = driver.find_element_by_xpath('//*[@id="'+ x +'"]/div[1]/section[1]/div/div')
occupation = occupation_element.text
image_element = driver.find_element_by_xpath('//*[@id="'+ x +'"]/div[1]/div[1]/div/figure/img')
image = image_element.get_attribute("src")

profile = {"name" : name, "telephone" : telephone, "occupation" : occupation, "image": image}
profiles.append(profile)
```

```python
driver.get(webs[1])
driver2 = webdriver.Chrome('chromedriver.exe')
ids = driver.find_elements_by_xpath("//*[contains(@id,'post-')]")
occupation = driver.find_elements_by_xpath('//*[@id="post-11"]/div/div[2]/div')
for i in range(1,len(occupation)+1):
    job = occupation[i-1].text.split("\n")[0]
    names = range(1, len(ids[0].find_elements_by_xpath('//*[@id="post-11"]/div/div[2]/div['+str(i)+']/div'))+1)
    for k in names:
        name = ids[0].find_elements_by_xpath('//*[@id="post-11"]/div/div[2]/div['+str(i)+']/div['+str(k)+']')[0].text.split("\n")[0]
        webpage = ids[0].find_elements_by_xpath('//*[@id="post-11"]/div/div[2]/div['+str(i)+']/div['+str(k)+']/a')[0].get_attribute("href")
        driver2.get(webpage)
        phone_element = driver2.find_elements_by_xpath('//*[@id="menu"]/p[2]/a')[0]
        phone = phone_element.text.split("\n")[0]
        image_element = driver2.find_elements_by_xpath('//*[@id="entries-box"]/div[1]/img')[0]
        image = image_element.get_attribute("src")
        profile = {"name" : name, "telephone" : phone, "occupation" : job, "image": image}
        profiles.append(profile)

```

```python
driver.get(webs[2])

name_element = driver.find_elements_by_xpath('/html/head/title')[0]
name = name_element.get_attribute("text").split("|")[0]
telephone_element = driver.find_elements_by_xpath('//*[@id="block-entityview-herocontent"]/div[1]/div[2]/div[2]/div/ul/li[4]')[0]
telephone = telephone_element.text.split("\n")[1]
occupation_element = driver.find_element_by_xpath('//*[@id="block-entityview-herocontent"]/div[1]/div[2]/div[2]/div/ul/li[3]/span[2]')
occupation = occupation_element.text
image_element = driver.find_element_by_xpath('//*[@id="block-entityview-herocontent"]/div[1]/div[1]/img')
image = image_element.get_attribute("src")

profile = {"name" : name, "telephone" : telephone, "occupation" : occupation, "image": image}
profiles.append(profile)
```
